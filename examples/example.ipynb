## Paulaner Wave

They can be used just as any other Matplotlib colormap. Here as an example of the classical Paulaner wave:

```python
cmaps = ["spezi", "orange", "zitrone"]
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.dpi"] = 150.
```

```python
xx, yy = np.meshgrid(np.linspace(0., 4.*np.pi, 100), np.linspace(0., 4.*np.pi, 100))
X = np.sin(xx) + yy - 0.2*xx
```

```python
fig, ax = plt.subplots(nrows=1, ncols=len(cmaps), figsize=(len(cmaps)*3.2, 3.2))
for i, cmap in enumerate(cmaps):
    ax[i].set(aspect=1., title=cmap)
    ax[i].contourf(xx, yy, X, cmap=paulaner_colormaps.__dict__[cmap], levels=4)
    ax[i].axis("off")
fig.set_layout_engine("tight")
fig.savefig("wave.png", bbox_inches="tight")
```

## Protoplanetary Disk

And here as an scientific example of the protoplanetary disk HD 163296 taken from [Andrews et al. (2018)](https://iopscience.iop.org/article/10.3847/2041-8213/aaf741). The example needs the [Astropy](https://www.astropy.org/) pacakge.

```python
from astropy.io import fits
```

```python
hdul = fits.open("https://almascience.eso.org/almadata/lp/DSHARP/images/HD163296_continuum.fits")
data = hdul[0].data[0, 0, 1000:2000, 1000:2000]
```

```python
X = np.clip(data, a_min=0., a_max=0.75*data.max())
```

```python
fig, ax = plt.subplots(nrows=1, ncols=len(cmaps), figsize=(len(cmaps)*3.2, 3.2))
for i, cmap in enumerate(cmaps):
    ax[i].set(aspect=1., title=cmap)
    ax[i].contourf(X, cmap=paulaner_colormaps.__dict__[cmap], levels=100)
    ax[i].axis("off")
fig.set_layout_engine("tight")
fig.savefig("hd163296.png", bbox_inches="tight")
```
